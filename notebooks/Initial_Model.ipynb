## **Data Cleaning**

```python
print("Cleaning Data (Removing Ocean/Empty Zones)...")
before_count = len(df)

# Filter: Must have either roads, commercial activity, or residential activity
valid_mask = (
    (df.get('road_density_total', 0) > 0.01) | 
    (df.get('poi_density_commercial', 0) > 0) | 
    (df.get('poi_density_residential', 0) > 0)
)
df = df[valid_mask].copy()
after_count = len(df)

print(f"Cleaning Complete.")
print(f"   -> Removed: {before_count - after_count:,} empty locations.")
print(f"   -> Remaining Valid Land Points: {after_count:,}")
```

## **Feature Engineering**

```python
print("Engineering Advanced Features...")

# 1. Calculate Land Use Entropy (Mixed-Use Indicator)
poi_cols = ['poi_density_commercial', 'poi_density_residential', 'poi_density_industrial', 
            'poi_density_recreational', 'poi_density_transport_hub']
# Ensure columns exist
existing_poi_cols = [c for c in poi_cols if c in df.columns]

if existing_poi_cols:
    df['total_poi'] = df[existing_poi_cols].sum(axis=1) + 1e-5 
    proportions = df[existing_poi_cols].div(df['total_poi'], axis=0)
    df['land_use_entropy'] = - (proportions * np.log(proportions + 1e-10)).sum(axis=1)
    print("   -> Created 'land_use_entropy'")
else:
    df['land_use_entropy'] = 0

# 2. Define Analysis Columns
analysis_cols = [
    'demand_score_kwh_only', 'land_use_entropy', 'poi_density_commercial', 
    'poi_density_residential', 'road_density_primary', 'urban_centrality_score', 
    'mixed_use_index'
]
analysis_cols = [c for c in analysis_cols if c in df.columns]

print(f"Feature Engineering Complete. Analysis columns: {len(analysis_cols)}")
```

## **Model Training**

```python
print("Training Ensemble AI Model...")

X = df[valid_feats].fillna(0)
# Create 3 Classes (Low, Med, High) based on Quantiles
y_class = pd.qcut(df['demand_score_kwh_only'].fillna(0), q=3, labels=[0, 1, 2])

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Define Models
clf1 = xgb.XGBClassifier(objective='multi:softmax', num_class=3, n_estimators=100, max_depth=6, n_jobs=-1, random_state=42)
clf2 = RandomForestClassifier(n_estimators=100, max_depth=8, n_jobs=-1, random_state=42)

# Ensemble
model = VotingClassifier(estimators=[('xgb', clf1), ('rf', clf2)], voting='soft')
model.fit(X_train, y_train)

# Validate
acc = model.score(X_test, y_test)
print(f"Model Training Complete.")
print(f"   -> Ensemble Validation Accuracy: {acc:.2%}")
print(f"   -> Baseline (Random) Accuracy: 33.33%")
print(f"   -> Improvement: +{acc - 0.3333:.2%}")
```

## **MCDM DECISION MAKING & RANKING**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
import folium
from folium.plugins import HeatMap

print("STARTING PHASE: MCDM RANKING...")

# 1. PREPARE INFERENCE DATA
# Ensure we drop columns that model doesn't expect (like IDs or Coordinates)
# NOTE: 'loaded_model' must be defined from the previous step
model_features = [
    'land_use_entropy', 
    'poi_density_commercial', 
    'poi_density_residential', 
    'road_density_primary', 
    'urban_centrality_score', 
    'mixed_use_index'
]
cols_to_drop = ['grid_id', 'latitude_x', 'longitude_x', 'geometry', 'grid_type_x', 'grid_type_y'] # Add your specific columns here
X_inference = data.drop(columns=[c for c in cols_to_drop if c in data.columns], errors='ignore')
X_inference = X_inference[model_features]
# Handle the text columns issue we discussed earlier
X_inference = X_inference.select_dtypes(exclude=['object'])
print(f"   -> Running predictions on {len(X_inference)} grid cells...")
preds = loaded_model.predict(X_inference)

# 2. SETUP CANDIDATES
candidates = data[['grid_id', 'latitude_x', 'longitude_x']].copy()
# Assign the raw predicted values directly
candidates['Predicted_Demand'] = preds 

# Use this predicted value as your metric for ranking
candidates['Demand_Metric'] = candidates['Predicted_Demand']

# 3. APPLY FILTERS (The Ocean/Wilderness Filter)
# We bring back infrastructure columns to ensure we don't build in the middle of nowhere
infra_cols = ['road_density_secondary', 'road_density_tertiary', 
              'poi_density_commercial', 'poi_density_residential', 'land_use_entropy']
# Join safely
candidates = candidates.join(data[[c for c in infra_cols if c in data.columns]].fillna(0))

# Calculate Local Infrastructure
candidates['Local_Road_Density'] = candidates.get('road_density_secondary', 0) + candidates.get('road_density_tertiary', 0)
candidates['Inhabited_Activity'] = candidates.get('poi_density_commercial', 0) + candidates.get('poi_density_residential', 0)

# Filter Logic
CONFIDENCE_THRESHOLD = 0.50
MIN_LOCAL_ROAD = 0.01
MIN_ACTIVITY = 0.05

high_confidence_sites = candidates[
    (candidates['Demand_Metric'] > CONFIDENCE_THRESHOLD) &
    (candidates['Local_Road_Density'] > MIN_LOCAL_ROAD) &
    (candidates['Inhabited_Activity'] > MIN_ACTIVITY)
].copy()

print(f"   -> Candidates passing filters: {len(high_confidence_sites)}")

if len(high_confidence_sites) == 0:
    print("CRITICAL WARNING: No candidates found. Lowering thresholds to force output...")
    # Fallback: Just take top 100 by demand if filters fail
    high_confidence_sites = candidates.sort_values(by='Demand_Metric', ascending=False).head(100).copy()

# 4. MCDM SCORING (INCLUDING ENTROPY)
print("   -> Calculating Multi-Criteria Scores...")
scaler = MinMaxScaler()

# Normalize Features (Handle missing columns safely)
norm_demand = scaler.fit_transform(high_confidence_sites[['Demand_Metric']])
norm_access = scaler.fit_transform(high_confidence_sites[['Local_Road_Density']])

# Check if we have Centrality and Entropy
c_score = data.loc[high_confidence_sites.index, 'urban_centrality_score'] if 'urban_centrality_score' in data.columns else 0
norm_centrality = scaler.fit_transform(c_score.values.reshape(-1, 1)) if 'urban_centrality_score' in data.columns else 0

e_score = data.loc[high_confidence_sites.index, 'land_use_entropy'] if 'land_use_entropy' in data.columns else 0
norm_entropy = scaler.fit_transform(e_score.values.reshape(-1, 1)) if 'land_use_entropy' in data.columns else 0

# --- WEIGHTED FORMULA (The Strategy) ---
# Demand: 40% (AI Prediction)
# Access: 20% (Roads)
# Centrality: 20% (Safety/Visibility)
# Entropy: 20% (Mixed-Use/Amenities - RESEARCH FEATURE)
high_confidence_sites['MCDM_Score'] = (
    (0.40 * norm_demand.flatten()) +
    (0.20 * norm_access.flatten()) +
    (0.20 * norm_centrality.flatten() if hasattr(norm_centrality, 'flatten') else 0) +
    (0.20 * norm_entropy.flatten() if hasattr(norm_entropy, 'flatten') else 0)
)

# 5. SPATIAL CLUSTERING (DEDUPLICATION)
print("   -> Running DBSCAN to group nearby candidates...")
coords = high_confidence_sites[['latitude_x', 'longitude_x']].values
db = DBSCAN(eps=0.005, min_samples=1, metric='euclidean').fit(coords)
high_confidence_sites['Cluster_ID'] = db.labels_

# 6. SELECT WINNERS (Top 1 per Cluster)
final_selection = []
for cluster_id, group in high_confidence_sites.groupby('Cluster_ID'):
    winner = group.loc[group['MCDM_Score'].idxmax()]
    final_selection.append(winner)

final_df = pd.DataFrame(final_selection).sort_values(by='MCDM_Score', ascending=False).head(20)

print("\nTOP 300 LOCATIONS SELECTED")
print(final_df[['grid_id', 'MCDM_Score', 'Predicted_Demand']].head())
```
